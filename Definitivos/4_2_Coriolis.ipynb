# Task 4.2 — Coriolis Force Effect

**Classical Mechanics · Project I · EETAC, UPC**

Compares the trajectory with Earth's rotation (ω = 7.27×10⁻⁵ rad/s) vs. no rotation (ω = 0).
Answers: landing displacement, most-affected direction, and physical explanation.

```python
# ── Download simulation dependencies from GitHub ─────────────────────────────
import urllib.request, os
BASE = "https://raw.githubusercontent.com/gdlcjoel-lab/PROYECTOMECANICA/main/Definitivos/"
for f in ["rocket_simulation.py", "plot_utils.py"]:
    if not os.path.exists(f):
        urllib.request.urlretrieve(BASE + f, f)
        print(f"Downloaded {f}")

# Use non-interactive backend so plots render inline in Colab
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
%matplotlib inline
print("Setup complete ✓")
```

```python
import numpy as np
from rocket_simulation import simulate_rocket, compute_metrics, print_metrics, OMEGA_EARTH

OMEGA_CASES = [OMEGA_EARTH, 0.0]
LABELS      = ["Coriolis ON (base)", "Coriolis OFF (omega=0)"]
COLORS      = ["#0D2B55", "#B07D2A"]
LINESTYLES  = ["--", "-"]

print("=" * 60)
print("  TASK 4.2 — CORIOLIS FORCE REMOVAL")
print("=" * 60)

results, metrics_list = [], []
for omega, label in zip(OMEGA_CASES, LABELS):
    print(f"\n>>> {label}  (omega = {omega:.2e} rad/s)")
    t, r, v, drag, mass, accel = simulate_rocket(omega=omega)
    m_obj = compute_metrics(t, r, v)
    results.append((t, r, v, drag, mass, accel))
    metrics_list.append(m_obj)
    print_metrics(m_obj, label)

m_on, m_off = metrics_list
print("\n1. HOW DOES THE TRAJECTORY CHANGE?")
print(f"   Coriolis ON:  x={m_on['x_land']:+.4f} m  y={m_on['y_land']:+.2e} m")
print(f"   Coriolis OFF: x={m_off['x_land']:+.4f} m  y={m_off['y_land']:+.2e} m")
print("\n2. WHICH DIRECTION IS MOST AFFECTED?")
print("   East-West (x): a_Cx = -2*omega*vz (deflects west on ascent)")
print(f"   Net west deflection: {abs(m_on['x_land']):.4f} m")
print(f"   North deviation:     {abs(m_on['y_land']):.2e} m  (effectively zero)")
```

## Plots
Altitude, East/North displacement and horizontal projection for both cases.

```python
t0, r0 = results[0][0], results[0][1]
t1, r1 = results[1][0], results[1][1]

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
fig.suptitle("Task 4.2 — Coriolis Effect ON vs OFF", fontsize=14, fontweight="bold", color="#0D2B55")
fig.patch.set_facecolor("white")
axes = axes.flatten()

for ax, x0, y0, x1, y1, title, xl, yl in [
    (axes[0], t0, r0[:,2], t1, r1[:,2], "Altitude vs Time", "Time [s]", "Altitude [m]"),
    (axes[1], t0, r0[:,0], t1, r1[:,0], "East Displacement vs Time", "Time [s]", "East x [m]"),
    (axes[2], t0, r0[:,1], t1, r1[:,1], "North Displacement vs Time", "Time [s]", "North y [m]"),
    (axes[3], r0[:,0], r0[:,1], r1[:,0], r1[:,1], "Horizontal Projection", "East x [m]", "North y [m]"),
]:
    ax.plot(x0, y0, color="#0D2B55", linestyle="--", linewidth=2, label=LABELS[0])
    ax.plot(x1, y1, color="#B07D2A", linestyle="-",  linewidth=2, label=LABELS[1])
    ax.set_title(title, fontweight="bold", color="#0D2B55")
    ax.set_xlabel(xl); ax.set_ylabel(yl)
    ax.legend(fontsize=9); ax.grid(True, alpha=0.4)

axes[3].set_aspect("equal", adjustable="datalim")
plt.tight_layout()
plt.savefig("task_4_2_plots.png", dpi=150, bbox_inches="tight")
plt.show()
print("Figure saved as task_4_2_plots.png")
```
